```python
# dropping unimportant rows
df1=df.drop(['url', 'address','menu_item','phone','dish_liked','reviews_list'],axis=1)
df1.head()
```

```python
df1.shape
```

```python
#checking null values
df1.info()
```

```python
#checking null values
df1.isnull().sum()
```

```python
#removing duplicate values
df1.drop_duplicates()
```

```python
#data cleaning
df1["rate"].unique()
```

```python
def handlerate(value):
    if(value=="NEW" or value=='-'):
        return np.nan  #returning null value in place of them
    else:
        value=str(value).split("/")
        value=value[0]
        return float(value)
df1["rate"]=df1["rate"].apply(handlerate)
df1["rate"].head()
```

```python
df1.rate.isnull().sum()
```

```python
#filling the null value in rate column with column mean
df1["rate"].fillna(df1["rate"].mean(),inplace=True)
df1.rate.isnull().sum()
```

```python
df1.isnull().sum()
```

```python
#dropping remaining rows null values
df1.dropna()
```

```python
df1.columns
```

```python
#rename the columns
df1=df1.rename(columns={"approx_cost(for two people)":"Cost2people","listed_in(type)":"Type"})
df1.head()
```

```python
df1.location.unique()
```

```python
df1["listed_in(city)"].unique()
```

```python
df2=df1.drop(["listed_in(city)"],axis=1)
df2.columns
```

```python
df2.head()
```

```python
#cleaning votes column
df2.votes.unique()
```

```python
#cleaning cost2people column
df2.Cost2people.unique()
```

```python
def handlecomma(value):
    value=str(value)
    if "," in value:
        value=value.replace(",","")
        return float(value)
    else:
        return float(value)
df2.Cost2people= df2.Cost2people.apply(handlecomma)
```

```python
df2.Cost2people.unique()
```

```python
#cleaning rest_type column
df2.rest_type.unique()
```

```python
rest_types=df2.rest_type.value_counts()
rest_types
```

```python
rest_types_lessthan1000=rest_types[rest_types<1000]
rest_types_lessthan1000
```

```python
def handle_rest_types(value):
    if(value in rest_types_lessthan1000):
        return 'others'
    else:
        return value
df2["rest_type"]=df2["rest_type"].apply(handle_rest_types)
df2["rest_type"].value_counts()
```

```python
df2.head()
```

```python
location=df2.location.value_counts()
location
```

```python
location_lessthan300=location[location<300]
```

```python
def handle_location(value):
    if(value in location_lessthan300):
        return 'others'
    else:
        return value
df2["location"]=df2["location"].apply(handle_location)
df2["location"].value_counts()
```

```python
df2.head()
```

```python
# cleaning cusinies column
cuisines=df2.cuisines.value_counts()
cuisines
```

```python
cuisines_lessthan100=cuisines[cuisines<100]
```

```python
def handle_cuisines(value):
    if(value in cuisines_lessthan100):
        return 'others'
    else:
        return value
df2["cuisines"]=df2["cuisines"].apply(handle_cuisines)
df2["cuisines"].value_counts()
```
