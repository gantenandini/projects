```python
import os
from pinecone import Pinecone, ServerlessSpec
index_name="hybrid-search-langchain-pinecone"

#initialize pinecone client
pc=Pinecone(api_key=api_key)

if index_name not in pc.list_indexes().names():
  pc.create_index(
      name=index_name,
      dimension=384, #dimension of dense vector
      metric='dotproduct',
      spec=ServerlessSpec(cloud='aws',region="us-east-1"),
  )
```

```python
index=pc.Index(index_name)
index
```

```python
from langchain_huggingface import HuggingFaceEmbeddings
embeddings=HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")
```

```python
embeddings
```

```python
from pinecone_text.sparse import BM25Encoder

bm25_encoder=BM25Encoder().default()
bm25_encoder
```

```python
sentences=["I ate dosa in morning","I ate rice in the afternoon","I ate roti at night"]
bm25_encoder.fit(sentences) #applying tf-idf values on these sentences

#store the values on to the json file

bm25_encoder.dump("bm25_values.json")
```

```python
retriever=PineconeHybridSearchRetriever(embeddings=embeddings,sparse_encoder=bm25_encoder,index=index)
```

```python
retriever
```

```python
retriever.add_texts(["I ate dosa in morning","I ate rice in the afternoon","I ate roti at night"])
```
